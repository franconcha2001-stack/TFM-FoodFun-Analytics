```python
df["Fecha Compra"] = df["Fecha Compra"].str.extract(
    r"(\d{1,2}\s+\w+,\s+\d{4})"
)[0]
```

```python
df["Fecha Compra"].head()
```

```python
meses = {
    "enero": "01",
    "febrero": "02",
    "marzo": "03",
    "abril": "04",
    "mayo": "05",
    "junio": "06",
    "julio": "07",
    "agosto": "08",
    "septiembre": "09",
    "octubre": "10",
    "noviembre": "11",
    "diciembre": "12"
}
```

```python
import re

def convertir_fecha(fecha):

    fecha = str(fecha).lower()

    patron = r"(\d{1,2})\s+(\w+),\s+(\d{4})"

    match = re.match(patron, fecha)

    if match:
        dia, mes, anio = match.groups()
        return f"{dia.zfill(2)}/{meses[mes]}/{anio}"

    return None

df["Fecha Compra"] = df["Fecha Compra"].apply(convertir_fecha)
```
